```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Insira seus dados para cada grupo etário
groups = {
    'A (>5 kg)': [1.2, 1.5, 1.7, 2.0, 2.3, 3.1, 4.0],  #
    'B (5<15 kg)': [2.1, 2.5, 2.8, 3.2, 3.9, 4.2, 5.1],
    'C (15<30 kg)': [5.2, 6.1, 6.5, 7.0, 7.8, 8.3, 9.1],
    'D (30<50 kg)': [10.0, 11.2, 12.5, 13.0, 14.3, 15.6],
    'E (50<80 kg)': [20.5, 21.2, 22.3, 23.8, 25.4],
    'F (>80 kg)': [30.5, 31.2, 33.3, 35.0, 40.0],
}

# Configuração do layout dos gráficos
fig, axes = plt.subplots(3, 2, figsize=(12, 15))
axes = axes.flatten()

for i, (group, data) in enumerate(groups.items()):
    data = np.array(data)  # Garantir que os dados estejam em formato numpy array

    # Calcular a mediana
    median = np.median(data)

    # Estimador de densidade kernel (KDE)
    kde = gaussian_kde(data)
    x = np.linspace(min(data), max(data), 1000)
    y = kde(x)

    # Plotar o histograma
    axes[i].hist(data, bins=15, density=True, alpha=0.5, label='Histogram', color='gray', edgecolor='black')

    # Plotar a curva de densidade
    axes[i].plot(x, y, 'r-', label='Density Curve (KDE)')

    # Adicionar a linha vertical para a mediana
    axes[i].axvline(median, color='red', linestyle='--', label=f'Median: {median:.2f}')

    # Configurações do gráfico
    axes[i].set_title(f'PKA Distribution for Group {group}', fontsize=12)
    axes[i].set_xlabel('PKA (Gy·cm²)', fontsize=10)
    axes[i].set_ylabel('Density of Procedures', fontsize=10)
    axes[i].legend(fontsize=8)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados para o gráfico
pesos = ['<5 kg', '5 a <15 kg', '15 a <30 kg', '30 a <50 kg', '50 a <80 kg']
drl_peso = [0.08245, 0.04355, 0.1430, 0.189, 0.307]

idades = ['<1 ano', '1 a <5 anos', '5 a <10 anos', '10 a <15 anos', '15 anos']
drl_idade = [0.0669, 0.088, 0.1700, 0.312, 0.279]

# Ajustando as posições no eixo X para os dois conjuntos de dados
x_peso = np.arange(len(pesos))
x_idade = np.arange(len(idades)) + len(pesos) + 1  # Separação entre os grupos

# Criar o gráfico
plt.figure(figsize=(8, 4))

# Adicionar barras para os dados de peso
bars_peso = plt.bar(x_peso, drl_peso, width=0.4, label='Por Peso (Gycm²)', color='skyblue', edgecolor='black')

# Adicionar barras para os dados de idade
bars_idade = plt.bar(x_idade, drl_idade, width=0.4, label='Por Idade (Gycm²)', color='salmon', edgecolor='black')

# Adicionar valores no topo de cada barra
for bar in bars_peso:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 1, yval + 0.005, f'{yval:.3f}', ha='center', va='bottom', fontsize=7)

for bar in bars_idade:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 1, yval + 0.005, f'{yval:.3f}', ha='center', va='bottom', fontsize=7)

# Adicionar rótulos nos eixos
plt.xticks(np.concatenate([x_peso, x_idade]), pesos + idades, rotation=20, fontsize=7)
plt.ylabel('Q2 de PKA (Gycm²)')
plt.title('Comparação de VTD de procedimeto FGI de inserção de CVC - por Peso e Idade')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
