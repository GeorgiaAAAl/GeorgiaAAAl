```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
df = pd.read_excel('/content/drive/MyDrive/georgia/georgia.xlsx')
display(df)
```

```python
categoria_cont = df['CATEGORIA'].value_counts()
categoria_cont = categoria_cont.reindex(['A', 'B', 'C', 'D', 'E', 'F'])
display(categoria_cont)
```

```python
categoriasestat = df.groupby('CATEGORIA')['PKA DAP total (Gycm²)'].describe()
categorias_dados = categoriasestat[['min', '50%', '75%', 'max','std']]
display(categorias_dados)
```

```python
faixa_kg = {
    'A': '< 5 kg',
    'B': '5 < 15 kg',
    'C': '15 < 30 kg',
    'D': '30 < 50 kg',
    'E': '50 < 80 kg',
    'F': '  80 kg < '
}
```

```python
#com outliers
plt.figure(figsize=(10, 5), facecolor='white')
sns.boxplot(x='CATEGORIA', y='PKA DAP total (Gy.cm^2)', data=df, color='skyblue')
plt.title('PKA total (Gy.cm²) por faixa de peso (kg)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Categoria', fontsize=13)
plt.ylabel('PKA Total (Gy.cm²)', fontsize=13)
plt.ylim(bottom=-0.5)
plt.xticks(rotation=15)
plt.xticks(ticks=[0, 1, 2, 3, 4, 5], labels=['< 5 kg', '5 < 15 kg', '15 < 30 kg', '30 < 50 kg', '50 < 80 kg', '80 kg <']) # se você não especificou quais são as categorias no texto

medianas = categorias_dados['50%']
legenda_textos = [f'{faixa_kg[categoria]} = {mediana:.3f}' for categoria, mediana in medianas.items()] # se você não especificou quais são as categorias no texto
plt.legend(legenda_textos, loc='upper left', bbox_to_anchor=(1, 1), fontsize=12, title="Medianas", title_fontsize=13, handlelength=0)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5), facecolor='white')

# Criando o boxplot sem outliers
sns.boxplot(x='CATEGORIA', y='PKA DAP total (Gycm²)', data=df, showfliers=False, whis=2, color='skyblue')

# Título e rótulos dos eixos
plt.title('PKA total (Gy.cm²) por faixa de peso (kg)', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Peso (kg)', fontsize=16)  # Alterado para "Peso (kg)"
plt.ylabel('PKA Total (Gy.cm²)', fontsize=16)

# Ajuste dos rótulos do eixo x
plt.xticks(ticks=[0, 1, 2, 3, 4, 5],
           labels=['< 5 kg', '5 < 15 kg', '15 < 30 kg', '30 < 50 kg', '50 < 80 kg', '80 kg <'],
           fontsize=14,
           rotation=15)

# Ajuste do tamanho da fonte dos valores no eixo y
ax = plt.gca()  # Obtém o eixo atual
ax.tick_params(axis='x', labelsize=14)  # Tamanho da fonte do eixo x
ax.tick_params(axis='y', labelsize=14)  # Tamanho da fonte do eixo y

# Adicionando legendas
medianas = categorias_dados['50%']
legenda_textos = [f'{faixa_kg[categoria]} = {mediana:.3f}' for categoria, mediana in medianas.items()]
plt.legend(legenda_textos, loc='upper right', bbox_to_anchor=(0.25, 1), fontsize=10, title="Peso (kg)", title_fontsize=13, handlelength=0)

# Definindo limites fixos para o eixo y
plt.ylim(0, 20)  # Escala do eixo y definida manualmente de 0 a 20

plt.show()
```
