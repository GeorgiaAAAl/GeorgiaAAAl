```python
# Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do arquivo Excel
file_path = '/content/BOXPLOT DE PKA x PESO.xlsx'  # Caminho para o arquivo
df_excel = pd.read_excel(file_path, sheet_name='PESO')


# Limpar e preparar os dados
df_cleaned = df_excel[['Mass (kg)', 'PKA DAP total (Gy.cm^2)']]  # Colunas relevantes
df_cleaned.columns = ['Peso (kg)', 'PKA (Gy.cm²)']  # Renomeando as colunas
df_cleaned['Peso (kg)'] = pd.to_numeric(df_cleaned['Peso (kg)'], errors='coerce')  # Convertendo para numérico
df_cleaned['PKA (Gy.cm²)'] = pd.to_numeric(df_cleaned['PKA (Gy.cm²)'], errors='coerce')  # Convertendo para numérico

# Criando categorias para faixas de peso
df_cleaned['Peso (kg) Category'] = pd.cut(df_cleaned['Peso (kg)'],
                                           bins=[0, 5, 15, 30, 50, 80, 100],
                                           labels=['<5 kg', '5<15 kg', '15<30 kg', '30<50 kg', '50<80 kg', '>80 kg'])

# Remover os dados com valores ausentes
df_cleaned = df_cleaned.dropna(subset=['Peso (kg)', 'PKA (Gy.cm²)'])

# Gerar o boxplot
plt.figure(figsize=(14, 7))
plt.boxplot([df_cleaned[df_cleaned['Peso (kg) Category'] == '<5 kg']['PKA (Gy.cm²)'],
             df_cleaned[df_cleaned['Peso (kg) Category'] == '5<15 kg']['PKA (Gy.cm²)'],
             df_cleaned[df_cleaned['Peso (kg) Category'] == '15<30 kg']['PKA (Gy.cm²)'],
             df_cleaned[df_cleaned['Peso (kg) Category'] == '30<50 kg']['PKA (Gy.cm²)'],
             df_cleaned[df_cleaned['Peso (kg) Category'] == '50<80 kg']['PKA (Gy.cm²)'],
             df_cleaned[df_cleaned['Peso (kg) Category'] == '>80 kg']['PKA (Gy.cm²)']],
            labels=['<5 kg', '5<15 kg', '15<30 kg', '30<50 kg', '50<80 kg', '>80 kg'])

# Adicionar título e rótulos
plt.title('Boxplot de PKA por Faixa de Peso dos Pacientes Pediátricos', fontsize=16)
plt.ylabel('PKA (Gy.cm²)', fontsize=12)
plt.xlabel('Faixa de Peso (kg)', fontsize=12)

# Exibir o gráfico
plt.show()
```

```python
# Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do arquivo Excel
file_path = '/content/BOXPLOT DE PKA x PESO.xlsx'  # Caminho para o arquivo
df_excel = pd.read_excel(file_path, sheet_name='PESO')

# Limpar e preparar os dados
df_cleaned = df_excel[['Mass (kg)', 'PKA DAP total (Gy.cm^2)']]  # Colunas relevantes
df_cleaned.columns = ['Peso (kg)', 'PKA (Gy.cm²)']  # Renomeando as colunas
df_cleaned['Peso (kg)'] = pd.to_numeric(df_cleaned['Peso (kg)'], errors='coerce')  # Convertendo para numérico
df_cleaned['PKA (Gy.cm²)'] = pd.to_numeric(df_cleaned['PKA (Gy.cm²)'], errors='coerce')  # Convertendo para numérico

# Criando categorias para faixas de peso
df_cleaned['Peso (kg) Category'] = pd.cut(df_cleaned['Peso (kg)'],
                                           bins=[0, 5, 15, 30, 50, 80, 100],
                                           labels=['<5 kg', '5<15 kg', '15<30 kg', '30<50 kg', '50<80 kg', '>80 kg'])

# Remover os dados com valores ausentes
df_cleaned = df_cleaned.dropna(subset=['Peso (kg)', 'PKA (Gy.cm²)'])

# Listar os valores únicos de 'Peso (kg) Category' para verificar as categorias presentes
print(df_cleaned['Peso (kg) Category'].unique())

# Criar o boxplot
plt.figure(figsize=(14, 7))

# Criando a lista de dados para o boxplot, incluindo as faixas de peso
data_for_boxplot = [
    df_cleaned[df_cleaned['Peso (kg) Category'] == '<5 kg']['PKA (Gy.cm²)'],
    df_cleaned[df_cleaned['Peso (kg) Category'] == '5<15 kg']['PKA (Gy.cm²)'],
    df_cleaned[df_cleaned['Peso (kg) Category'] == '15<30 kg']['PKA (Gy.cm²)'],
    df_cleaned[df_cleaned['Peso (kg) Category'] == '30<50 kg']['PKA (Gy.cm²)'],
    df_cleaned[df_cleaned['Peso (kg) Category'] == '50<80 kg']['PKA (Gy.cm²)'],
    df_cleaned[df_cleaned['Peso (kg) Category'] == '>80 kg']['PKA (Gy.cm²)']
]

# Gerar o boxplot
plt.boxplot(data_for_boxplot, labels=['<5 kg', '5<15 kg', '15<30 kg', '30<50 kg', '50<80 kg', '>80 kg'])

# Adicionar título e rótulos
plt.title('Boxplot de PKA por Faixa de Peso dos Pacientes Pediátricos', fontsize=16)
plt.ylabel('PKA (Gy.cm²)', fontsize=12)
plt.xlabel('Faixa de Peso (kg)', fontsize=12)

# Exibir o gráfico
plt.show()
```
